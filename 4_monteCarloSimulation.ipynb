## 2. ÏãúÎÆ¨Î†àÏù¥ÏÖò 1Ìöå ÎèåÎ¶¨Îäî Ìï®ÏàòÏûëÏÑ±

### ‚ë† Î™©ÌëúÎß§Ï∂úÏï° Ï¥àÏù¥Ïä§ Î∂ÑÌè¨ ÎßåÎì§Í∏∞

```python
# scipyÎ•º ÏÇ¨Ïö©Ìïú ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ïù¥ÏÇ∞ ÌôïÎ•†Î∂ÑÌè¨ Í∏∞Î∞ò ÏÉòÌîåÎßÅ
from scipy import stats # scipy Ìå®ÌÇ§ÏßÄÏùò stats(statistics) Î™®Îìà ÏûÑÌè¨Ìä∏
import matplotlib.pyplot as plt


sales_targets = [75_100, 100_000, 200_000, 300_000, 400_000, 500_000]
sales_probabilities = [0.3, 0.3, 0.2, 0.1, 0.05, 0.05]
size = 500
# scipyÏùò rv_discrete(random variable / descrete(Ïù¥ÏÇ∞Ìòï))ÏùÑ Ïù¥Ïö©Ìï¥ ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ïù¥ÏÇ∞ Î∂ÑÌè¨ ÏÉùÏÑ±
target_sales_dist = stats.rv_discrete(name='targets_sales_dist', values=(sales_targets, sales_probabilities))

# Ï†ïÏùòÎêú Ïù¥ÏÇ∞Î∂ÑÌè¨Î•º Í∏∞Î∞òÏúºÎ°ú 500Î™ÖÏùò ÏÉòÌîå ÏÉùÏÑ±
target_sales_samples = target_sales_dist.rvs(size=size)

# ÏÉòÌîåÎßÅÎêú Î™©Ìëú Îß§Ï∂ú Î∂ÑÌè¨Î•º ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú ÏãúÍ∞ÅÌôî
plt.hist(x = target_sales_samples, bins=10)
plt.show()
```

### ‚ë° Î™©ÌëúÎã¨ÏÑ±Î•† Ï†ïÍ∑úÎ∂ÑÌè¨ ÎßåÎì§Í∏∞

```python
# loc ‚Üí location parameter (ÏúÑÏπòÎ™®Ïàò) ‚Üí Î∂ÑÌè¨Ïùò Ï§ëÏã¨ (Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò Í≤ΩÏö∞ ÌèâÍ∑† Œº)
# scale ‚Üí scale parameter (Ï≤ôÎèÑÎ™®Ïàò) ‚Üí Î∂ÑÌè¨Ïùò ÌçºÏßê (Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò Í≤ΩÏö∞ ÌëúÏ§ÄÌé∏Ï∞® œÉ)
# ÏÉùÏÑ±Îêú Í∞íÏùÄ ÏÜåÏàòÏ†ê ÎëòÏß∏ ÏûêÎ¶¨ÍπåÏßÄ Î∞òÏò¨Î¶º
np.random.seed(0)
achievement_rates_samples = np.random.normal(loc = 1, scale = 0.1, size = size).round(2)


# ÏÉùÏÑ±Îêú Î™©Ìëú Îã¨ÏÑ±Î•†ÏùÑ ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú ÏãúÍ∞ÅÌôî
plt.hist(achievement_rates_samples, bins=10)
plt.show()
```

### ‚ë¢ 1Í∞ú ÏãúÎÆ¨Î†àÏù¥ÏÖòÏóê ÎåÄÌïú Î∞úÏÉùÎß§Ï∂úÏï° Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÎßåÎì§Í∏∞

```python
# ÏÉùÏÑ±Ìïú Î™©ÌëúÎß§Ï∂úÏï°ÏÉòÌîåÍ≥º Î™©ÌëúÎã¨ÏÑ±Î•†ÏÉòÌîåÏùÑ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Í≤∞Ìï©
df = pd.DataFrame(
    index=range(size),
    data={
        'target_sales': target_sales_samples,  # Î™©ÌëúÎß§Ï∂úÏï°500Î™ÖÏÉòÌîå
        'achievement_rates': achievement_rates_samples  # Î™©ÌëúÎã¨ÏÑ±Î•†500Î™ÖÏÉòÌîå
    }
)

# Î∞úÏÉùÎß§Ï∂úÏï° = Î™©ÌëúÎß§Ï∂úÏï° √ó Î™©ÌëúÎã¨ÏÑ±Î•†
df['actual_sales'] = df['target_sales'] * df['achievement_rates']

# Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ï∂úÎ†•
df
```

### ‚ë£ ÌåêÎß§ÏàòÏàòÎ£åÏú® Î∞è ÌåêÎß§ ÏàòÏàòÎ£å ÌïÑÎìú Î∂ôÏù¥Í∏∞

```python
def commission_rate(x):
    if x <= 0.9:
        return 0.02
    elif x <= 0.99:
        return 0.03
    else:
        return 0.04

# Í∞Å ÌñâÏùò Î™©Ìëú Îã¨ÏÑ±Î•†Ïóê Îî∞Îùº ÏàòÏàòÎ£åÏú® Í≥ÑÏÇ∞
df['commission_rate'] = df['achievement_rates'].apply(commission_rate)

# Ïã§Ï†ú Îß§Ï∂ú √ó ÏàòÏàòÎ£åÏú® ‚Üí ÌåêÎß§ ÏàòÏàòÎ£å Í≥ÑÏÇ∞
df['commission_amount'] = df['actual_sales'] * df['commission_rate']

# Í≤∞Í≥º Ï∂úÎ†•
print(df)
```

### ‚ë§ ÏÉÅÍ∏∞Í≥ºÏ†ïÏùÑ Ìï®ÏàòÎ°ú ÏûëÏÑ±

```python
def create_sales_simulation(size=500,
                            sales_targets = [75_100, 100_000, 200_000, 300_000, 400_000, 500_000], 
                            sales_probabilities = [0.3, 0.3, 0.2, 0.1, 0.05, 0.05],
                            ):
    # Î™©ÌëúÎß§Ï∂úÏï° Ïù¥ÏÇ∞Î∂ÑÌè¨ ÎßåÎì§Í∏∞
    target_sales_dist = stats.rv_discrete(name='targets_sales_dist', values=(sales_targets, sales_probabilities))
    target_sales_samples = target_sales_dist.rvs(size=size)

    # Î™©ÌëúÎã¨ÏÑ±Î•† Ï†ïÍ∑úÎ∂ÑÌè¨ ÎßåÎì§Í∏∞
    achievement_rates_samples = np.random.normal(loc=1, scale=0.1, size=size).round(2)

    # 1Í∞ú ÏãúÎÆ¨Î†àÏù¥ÏÖòÏóê ÎåÄÌïú Î∞úÏÉùÎß§Ï∂úÏï° Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÎßåÎì§Í∏∞
    df = pd.DataFrame(
        index=range(size),
        data={
            'target_sales': target_sales_samples,  # Î™©ÌëúÎß§Ï∂úÏï°ÏÉòÌîå
            'achievement_rates': achievement_rates_samples  # Î™©ÌëúÎã¨ÏÑ±Î•†ÏÉòÌîå
        }
    )

    # Î∞úÏÉùÎß§Ï∂úÏï° = Î™©ÌëúÎß§Ï∂úÏï° √ó Î™©ÌëúÎã¨ÏÑ±Î•†
    df['actual_sales'] = df['target_sales'] * df['achievement_rates']

    # ÌåêÎß§ÏàòÏàòÎ£åÏú® Î∞è ÌåêÎß§ ÏàòÏàòÎ£å ÌïÑÎìú Î∂ôÏù¥Í∏∞
    def commission_rate(x):
        if x <= 0.9:
            return 0.02
        elif x <= 0.99:
            return 0.03
        else:
            return 0.04

    # Í∞Å ÌñâÏùò Î™©Ìëú Îã¨ÏÑ±Î•†Ïóê Îî∞Îùº ÏàòÏàòÎ£åÏú® Í≥ÑÏÇ∞
    df['commission_rate'] = df['achievement_rates'].apply(commission_rate)

    # Ïã§Ï†ú Îß§Ï∂ú √ó ÏàòÏàòÎ£åÏú® ‚Üí ÌåêÎß§ ÏàòÏàòÎ£å Í≥ÑÏÇ∞
    df['commission_amount'] = df['actual_sales'] * df['commission_rate']

    return df

create_sales_simulation()
```

```python
# ÏãúÎÆ¨Î†àÏù¥ÏÖò Í≤∞Í≥ºÎ•º Ï†ÄÏû•Ìï† Îπà Î¶¨Ïä§Ìä∏ ÏÉùÏÑ±
all_stats = []
# ÏãúÎÆ¨Î†àÏù¥ÏÖò Î∞òÎ≥µ ÌöåÏàò ÏßÄÏ†ï
num_simulations = 1000
# ÎÇúÏàò ÏãúÎìú Ï¥àÍ∏∞Ìôî (ÏãúÎÆ¨Î†àÏù¥ÏÖò Í≤∞Í≥º Ïû¨ÌòÑ Í∞ÄÎä•ÌïòÎèÑÎ°ù)
np.random.seed(0)

# ÏãúÎÆ¨Î†àÏù¥ÏÖò Î∞òÎ≥µ ÏàòÌñâ ÌïòÏó¨ Î¶¨Ïä§Ìä∏Ïóê Ï†ÅÏû¨ÌïòÍ∏∞
for i in range(num_simulations):
    df = create_sales_simulation()

    # Í∞Å ÏãúÎÆ¨Î†àÏù¥ÏÖòÎ≥Ñ Ï¥ùÌï© Í≥ÑÏÇ∞ÌïòÏó¨ Î¶¨Ïä§Ìä∏Ïóê Ï†ÄÏû•
    all_stats.append([
        df['target_sales'].sum().round(0),        # Ï¥ù Î™©Ìëú Îß§Ï∂ú
        df['actual_sales'].sum().round(0),        # Ï¥ù Î∞úÏÉù Îß§Ï∂ú
        df['commission_amount'].sum().round(0)    # Ï¥ù ÏàòÏàòÎ£å
    ])

# Î¶¨Ïä§Ìä∏Ïóê Ï†ÅÏû¨Îêú ÏãúÎÆ¨Î†àÏù¥ÏÖò Í≤∞Í≥ºÎ•º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Î≥ÄÌôò
results_df = pd.DataFrame.from_records(
    all_stats,
    columns=['Total Target Sales', 'Total Actual Sales', 'Total Commissions Amount']
)

results_df
```

# cf. NumpyÎ°ú Î™¨ÌÖåÏπ¥Î•ºÎ°ú ÏãúÎÆ¨Î†àÏù¥ÏÖò ÎèåÎ†§Î≥¥Í∏∞

```python
import numpy as np
import matplotlib.pyplot as plt

# ÏãúÎÆ¨Î†àÏù¥ÏÖò Í≤∞Í≥º(Ï¥ù ÏàòÏàòÎ£å Ìï©Í≥Ñ)Î•º Ï†ÄÏû•Ìï† Î¶¨Ïä§Ìä∏
total_commissions_all = []

# ÎÇúÏàò Í≥†Ï†ï (Ïû¨ÌòÑÏÑ± ÌôïÎ≥¥)
np.random.seed(0)

# ÏãúÎÆ¨Î†àÏù¥ÏÖò Î∞òÎ≥µ ÌöüÏàò
num_simulations = 1000

# Í∞Å ÏãúÎÆ¨Î†àÏù¥ÏÖòÏóêÏÑú ÏÉùÏÑ±Ìï† ÏÉòÌîå Ïàò (Ïòà: ÌïòÎ£® Í∏∞Ï§Ä 500Í∞ú)
num_trials = 500

# Î™©Ìëú Îß§Ï∂ú Íµ¨Í∞ÑÍ≥º Ìï¥Îãπ ÌôïÎ•†
sales_targets = [75_000, 100_000, 200_000, 300_000, 400_000, 500_000]
sales_probabilities = [0.3, 0.3, 0.2, 0.1, 0.05, 0.05]

# Î™©Ìëú Îã¨ÏÑ±Î•†Ïùò ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞®
mean = 1.0
std_dev = 0.1

for i in range(num_simulations):
    # Î™©Ìëú Îß§Ï∂ú ÏÉòÌîå ÏÉùÏÑ± (Ïù¥ÏÇ∞ ÌôïÎ•†Î∂ÑÌè¨ Í∏∞Î∞ò)
    target_sales = np.random.choice(sales_targets, num_trials, p=sales_probabilities)

    # Î™©Ìëú Îã¨ÏÑ±Î•† ÏÉòÌîå ÏÉùÏÑ± (Ï†ïÍ∑úÎ∂ÑÌè¨ Í∏∞Î∞ò)
    achievement_rate = np.random.normal(mean, std_dev, num_trials).round(2)

    # Ïã§Ï†ú Îß§Ï∂ú = Î™©Ìëú Îß§Ï∂ú √ó Îã¨ÏÑ±Î•†
    actual_sales = target_sales * achievement_rate

    # ÏàòÏàòÎ£åÏú® Í≥ÑÏÇ∞: Ï°∞Í±¥Ïóê Îî∞Îùº Îã§Î•∏ ÎπÑÏú® Ï†ÅÏö©
    #  - 90% Ïù¥Ìïò: 2%
    #  - 90% Ï¥àÍ≥º ~ 100% ÎØ∏Îßå: 3%
    #  - 100% Ïù¥ÏÉÅ: 4%
    commission_rate = np.where(achievement_rate <= 0.9, 0.02,
                        np.where(achievement_rate < 1.0, 0.03, 0.04))

    # ÏàòÏàòÎ£å Í∏àÏï° Í≥ÑÏÇ∞
    commission_amount = commission_rate * actual_sales

    # Ïù¥Î≤à ÏãúÎÆ¨Î†àÏù¥ÏÖòÏùò Ï†ÑÏ≤¥ ÏàòÏàòÎ£å Ìï©Í≥ÑÎ•º Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞Ä
    total_commissions_all.append(np.sum(commission_amount))

# Í≤∞Í≥ºÎ•º ÎÑòÌååÏù¥ Î∞∞Ïó¥Î°ú Î≥ÄÌôò
total_commissions_all = np.array(total_commissions_all)

# ÌûàÏä§ÌÜ†Í∑∏Îû® ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 5))
plt.hist(total_commissions_all, bins=100, edgecolor='black')
plt.axvline(x=3_000_000, color='red', linestyle='dashed', label='Budget: $3,000,000')
plt.title("Total Commissions Distribution")
plt.xlabel("Total Commissions")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# ÏàòÏàòÎ£å ÏòàÏÇ∞ ÏûÑÍ≥ÑÏπò ÏÑ§Ï†ï
threshold = 3_000_000

# ÏµúÏÜå, ÏµúÎåÄ ÏàòÏàòÎ£å Í∞í Í≥ÑÏÇ∞
min_val = np.min(total_commissions_all)
max_val = np.max(total_commissions_all)

# ÏûÑÍ≥ÑÏπòÎ•º Í∏∞Ï§ÄÏúºÎ°ú Îëê Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÏñ¥ ÌûàÏä§ÌÜ†Í∑∏Îû® Ïπ¥Ïö¥Ìä∏ Í≥ÑÏÇ∞
count, bins = np.histogram(total_commissions_all, bins=[min_val, threshold, max_val])

# Í∞Å Íµ¨Í∞ÑÏùò ÌôïÎ•†(PDF) Í≥ÑÏÇ∞
pdf = count / sum(count)
print("üìä PDF (ÌôïÎ•†):", pdf)
```
