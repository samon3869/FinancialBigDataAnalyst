# 🎲 Monte Carlo Simulation 설명


## 🔍 몬테카를로 시뮬레이션이란?

**몬테카를로 시뮬레이션(Monte Carlo Simulation)**은 확률적 입력값(난수)을 이용해 불확실한 상황 하에서의 리스크나 결과를 반복적으로 계산하여, 결과의 **분포와 변동성**, 그리고 **주요 영향을 미치는 요인**을 파악하는 방법입니다.  

즉, 미래의 결과가 여러 불확실한 변수에 따라 달라질 수 있을 때, 단 하나의 평균값에 의존하지 않고 **수천 번의 시뮬레이션**을 통해 다양한 가능성을 분석합니다.

## 💥 왜 평균값만으로는 부족한가?

많은 분석이 단일 평균값(예: "평균 수익률")만을 기준으로 판단하는데, 이는 다음과 같은 **인지 편향(Cognitive Bias)**으로 이어질 수 있습니다:

- 평균은 극단적인 상황(특히 **하방 리스크**)을 반영하지 못함
- 현실은 여러 변수들이 복합적으로 변동하며, 단일 시나리오로는 이를 포착하기 어려움
- 엑셀의 일반적 연산은 대부분 **하나의 값(정확한 수치)**만을 입력값으로 사용함

👉 이 때문에 **시나리오 분석**이나 **민감도 분석**이 필요하지만, 이들에도 한계가 있습니다.

## ⚙️ 민감도 분석 vs. 시뮬레이션 분석

| 분석 방법         | 개요                                                       | 한계점 |
|------------------|------------------------------------------------------------|--------|
| **민감도 분석**   | 변수 1개씩 일정 수준으로 변화시켜 결과를 비교              | 변화의 **확률**이나 **복합적 변수 간 상호작용**은 분석 불가 |
| **시나리오 분석** | 최악 / 보통 / 최선 시나리오를 만들어 비교                  | 현실의 확률분포를 반영하기 어려움 |
| **시뮬레이션 분석 (몬테카를로)** | 확률분포에 따라 난수 생성 → 수천 번 결과 계산 → 결과 통계 분석 | 변수의 분포 설정과 컴퓨터 계산 자원이 필요 |

## 🧠 몬테카를로 시뮬레이션이 리스크 분석에 강한 이유

리스크는 다음 수식으로 표현할 수 있습니다:

> **리스크 = 바람직하지 않은 결과 × 발생할 가능성(확률)**

이때 중요한 것은 평균값이 아니라 **결과의 산포와 하방 리스크**입니다.

→ 따라서 **분산, 표준편차, 퍼센타일, Value at Risk (VaR)** 등의 지표가 매우 중요해집니다.

**몬테카를로 시뮬레이션은 다음을 가능하게 합니다:**

1. **결과값 분포 시각화** (히스토그램, 신뢰구간 등)
2. **하방 리스크 정량화** (예: 최악의 5% 손실 등)
3. **민감도 분석 병행** → 결과에 큰 영향을 주는 변수 식별 가능

## 🏛️ 어원: "Monte Carlo"는 어디서 왔을까?

‘몬테카를로(Monte Carlo)’는 **모나코의 카지노 도시** 이름에서 유래되었습니다.  
→ 이 도시처럼 **무작위(random)성과 확률**이 중요한 요소이기 때문입니다.

이 기법은 제2차 세계대전 당시 **맨해튼 프로젝트**에서 **핵 확산을 시뮬레이션하기 위해 개발**되었고,  
**수학자 스태니스와프 울람(Stanislaw Ulam)**과 **존 폰 노이만(John von Neumann)**이 명명하고 체계화했습니다.

## 1. 해결할 문제
500명의 영업사원들에게 지급할 판매수수료 금액 예측하여 예산 설정하기

판매수수료 = 실제 매출(목표 매출액 및 목표달성률 분포에 종속) * 수수료율 (목표 매출액 및 목표달성률 분포에 종속)

---
##### 목표매출액 분포 (분포가정1 - 초이스분포)

|--------|------------|
|  75,000 | 30%        |
| 100,000 | 30%        |
| 200,000 | 20%        |
| 300,000 | 10%        |
| 400,000 | 5%         |

---

##### 목표달성률 분포 (분포가정2 - 정규분포)
| 구분  | 값 |
|--------|------------|
|  분포종류  | 정규분포  |

---

##### 목표달성률에 따른 수수료율

| 구분  | 값 |

## 2. 시뮬레이션 1회 돌리는 함수작성

### ① 목표매출액 초이스 분포 만들기

```python
# scipy를 사용한 사용자 정의 이산 확률분포 기반 샘플링
from scipy import stats # scipy 패키지의 stats(statistics) 모듈 임포트
import matplotlib.pyplot as plt


sales_targets = [75_100, 100_000, 200_000, 300_000, 400_000, 500_000]
sales_probabilities = [0.3, 0.3, 0.2, 0.1, 0.05, 0.05]
size = 500
# scipy의 rv_discrete(random variable / descrete(이산형))을 이용해 사용자 정의 이산 분포 생성
target_sales_dist = stats.rv_discrete(name='targets_sales_dist', values=(sales_targets, sales_probabilities))

# 정의된 이산분포를 기반으로 500명의 샘플 생성
target_sales_samples = target_sales_dist.rvs(size=size)

# 샘플링된 목표 매출 분포를 히스토그램으로 시각화
plt.hist(x = target_sales_samples, bins=10)
plt.show()
```

### ② 목표달성률 정규분포 만들기

```python
# loc → location parameter (위치모수) → 분포의 중심 (정규분포의 경우 평균 μ)
# scale → scale parameter (척도모수) → 분포의 퍼짐 (정규분포의 경우 표준편차 σ)
# 생성된 값은 소수점 둘째 자리까지 반올림
np.random.seed(0)
achievement_rates_samples = np.random.normal(loc = 1, scale = 0.1, size = size).round(2)


# 생성된 목표 달성률을 히스토그램으로 시각화
plt.hist(achievement_rates_samples, bins=10)
plt.show()
```

### ③ 1개 시뮬레이션에 대한 발생매출액 데이터프레임 만들기

```python
# 생성한 목표매출액샘플과 목표달성률샘플을 데이터프레임으로 결합
df = pd.DataFrame(
    index=range(size),
    data={
        'target_sales': target_sales_samples,  # 목표매출액500명샘플
        'achievement_rates': achievement_rates_samples  # 목표달성률500명샘플
    }
)

# 발생매출액 = 목표매출액 × 목표달성률
df['actual_sales'] = df['target_sales'] * df['achievement_rates']

# 결과 데이터프레임 출력
df
```

### ④ 판매수수료율 및 판매 수수료 필드 붙이기

```python
def commission_rate(x):
    if x <= 0.9:
        return 0.02
    elif x <= 0.99:
        return 0.03
    else:
        return 0.04

# 각 행의 목표 달성률에 따라 수수료율 계산
df['commission_rate'] = df['achievement_rates'].apply(commission_rate)

# 실제 매출 × 수수료율 → 판매 수수료 계산
df['commission_amount'] = df['actual_sales'] * df['commission_rate']

# 결과 출력
print(df)
```

### ⑤ 상기과정을 함수로 작성

```python
def create_sales_simulation(size=500,
                            sales_targets = [75_100, 100_000, 200_000, 300_000, 400_000, 500_000], 
                            sales_probabilities = [0.3, 0.3, 0.2, 0.1, 0.05, 0.05],
                            ):
    # 목표매출액 이산분포 만들기
    target_sales_dist = stats.rv_discrete(name='targets_sales_dist', values=(sales_targets, sales_probabilities))
    target_sales_samples = target_sales_dist.rvs(size=size)

    # 목표달성률 정규분포 만들기
    achievement_rates_samples = np.random.normal(loc=1, scale=0.1, size=size).round(2)

    # 1개 시뮬레이션에 대한 발생매출액 데이터프레임 만들기
    df = pd.DataFrame(
        index=range(size),
        data={
            'target_sales': target_sales_samples,  # 목표매출액샘플
            'achievement_rates': achievement_rates_samples  # 목표달성률샘플
        }
    )

    # 발생매출액 = 목표매출액 × 목표달성률
    df['actual_sales'] = df['target_sales'] * df['achievement_rates']

    # 판매수수료율 및 판매 수수료 필드 붙이기
    def commission_rate(x):
        if x <= 0.9:
            return 0.02
        elif x <= 0.99:
            return 0.03
        else:
            return 0.04

    # 각 행의 목표 달성률에 따라 수수료율 계산
    df['commission_rate'] = df['achievement_rates'].apply(commission_rate)

    # 실제 매출 × 수수료율 → 판매 수수료 계산
    df['commission_amount'] = df['actual_sales'] * df['commission_rate']

    return df

create_sales_simulation()
```

## 3. 시뮬레이션 1000번 돌리기

### ① 시뮬레이션 1000번 돌린 집계데이터를 데이터프레임에 저장

```python
# 시뮬레이션 결과를 저장할 빈 리스트 생성
all_stats = []
# 시뮬레이션 반복 회수 지정
num_simulations = 1000
# 난수 시드 초기화 (시뮬레이션 결과 재현 가능하도록)
np.random.seed(0)

# 시뮬레이션 반복 수행 하여 리스트에 적재하기
for i in range(num_simulations):
    df = create_sales_simulation()

    # 각 시뮬레이션별 총합 계산하여 리스트에 저장
    all_stats.append([
        df['target_sales'].sum().round(0),        # 총 목표 매출
        df['actual_sales'].sum().round(0),        # 총 발생 매출
        df['commission_amount'].sum().round(0)    # 총 수수료
    ])

# 리스트에 적재된 시뮬레이션 결과를 데이터프레임으로 변환
results_df = pd.DataFrame.from_records(
    all_stats,
    columns=['Total Target Sales', 'Total Actual Sales', 'Total Commissions Amount']
)

results_df
```

### ② 통계정보 출력

```python
# 통계 요약 출력 (소수점 8자리까지 반올림)
results_df.describe().round(8)
```

```python
# 수수료 총합의 분포를 히스토그램으로 시각화
plt.hist(results_df['Total Commissions Amount'], bins=10)
plt.show()
```

## 4. 경영진 질문 답하기

```python
# 예산이 3,000,000원일 때 몇 퍼센트 확률이 이 예사니 내년 판매수수료를 충당할 수 있을까? (예산 대비 판매수수료 총합이 1 이하일 확률)



# 판매 수수료 총합 결과에서 수치만 추출
commission_values = results_df['Total Commissions Amount']

# 수수료 예산 한도 설정
threshold = 3_000_000

# 수수료 총합의 최소값과 최대값 계산
min_val = np.min(commission_values)
max_val = np.max(commission_values)

# 수수료가 예산 이내인 경우와 초과한 경우를 구분하는 히스토그램 구간 설정
# [min_val, threshold), [threshold, max_val]
count, bins_values = np.histogram(commission_values, bins = [min_val, threshold, max_val])
print(count)
print(bins_values)
# 각 구간의 확률(PDF: 확률밀도함수) 계산
pdf = count / sum(count)  # 전체 시뮬레이션 횟수로 나누어 비율로 변환
print("📊 PDF (확률):", pdf)

# # 누적 확률(CDF: 누적분포함수) 계산
cdf = np.cumsum(pdf)
print("📈 CDF (누적확률):", cdf)
```

# cf. Numpy로 몬테카를로 시뮬레이션 돌려보기

```python
import numpy as np
import matplotlib.pyplot as plt

# 시뮬레이션 결과(총 수수료 합계)를 저장할 리스트
total_commissions_all = []

# 난수 고정 (재현성 확보)
np.random.seed(0)

# 시뮬레이션 반복 횟수
num_simulations = 1000

# 각 시뮬레이션에서 생성할 샘플 수 (예: 하루 기준 500개)
num_trials = 500

# 목표 매출 구간과 해당 확률
sales_targets = [75_000, 100_000, 200_000, 300_000, 400_000, 500_000]
sales_probabilities = [0.3, 0.3, 0.2, 0.1, 0.05, 0.05]

# 목표 달성률의 평균과 표준편차
mean = 1.0
std_dev = 0.1

for i in range(num_simulations):
    # 목표 매출 샘플 생성 (이산 확률분포 기반)
    target_sales = np.random.choice(sales_targets, num_trials, p=sales_probabilities)

    # 목표 달성률 샘플 생성 (정규분포 기반)
    achievement_rate = np.random.normal(mean, std_dev, num_trials).round(2)

    # 실제 매출 = 목표 매출 × 달성률
    actual_sales = target_sales * achievement_rate

    # 수수료율 계산: 조건에 따라 다른 비율 적용
    #  - 90% 이하: 2%
    #  - 90% 초과 ~ 100% 미만: 3%
    #  - 100% 이상: 4%
    commission_rate = np.where(achievement_rate <= 0.9, 0.02,
                        np.where(achievement_rate < 1.0, 0.03, 0.04))

    # 수수료 금액 계산
    commission_amount = commission_rate * actual_sales

    # 이번 시뮬레이션의 전체 수수료 합계를 리스트에 추가
    total_commissions_all.append(np.sum(commission_amount))

# 결과를 넘파이 배열로 변환
total_commissions_all = np.array(total_commissions_all)

# 히스토그램 시각화
plt.figure(figsize=(10, 5))
plt.hist(total_commissions_all, bins=100, edgecolor='black')
plt.axvline(x=3_000_000, color='red', linestyle='dashed', label='Budget: $3,000,000')
plt.title("Total Commissions Distribution")
plt.xlabel("Total Commissions")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# 수수료 예산 임계치 설정
threshold = 3_000_000

# 최소, 최대 수수료 값 계산
min_val = np.min(total_commissions_all)
max_val = np.max(total_commissions_all)

# 임계치를 기준으로 두 구간으로 나누어 히스토그램 카운트 계산
count, bins = np.histogram(total_commissions_all, bins=[min_val, threshold, max_val])

# 각 구간의 확률(PDF) 계산
pdf = count / sum(count)
print("📊 PDF (확률):", pdf)
```
