# ㈜한공 몬테카를로 시뮬레이션 문제

㈜한공은 이제까지 향후 매출액을 **결정론적 접근방식**에 따라 단일 평균값으로 추정하였다.  
그러나 이 방법으로는 순이익 예측 모형에서의 **불확실성**을 제대로 반영하지 못한다는 컨설팅 의견을 받았다.  

따라서 재무이사는 다음 회계연도의 **당기순이익(Net Income)** 에 대해 **몬테카를로 시뮬레이션**을 실행하고자 한다.  
이를 위해 먼저 **추정손익계산서**를 작성한다. (양식: `[추정손익계산서.tbl]`, 단위: 원)

---

## 입력 데이터

- **월별 순매출**:  
  - [분개장_1.csv] → pandas DataFrame 불러오기  
  - 간주계정과목이 "매출"인 계정의  
    `대변금액 합계 – 차변금액 합계 = 월별 순매출`

- **매출원가율**:  
  - 과거 100개월 자료: `[과거매출원가율.tbl]`  
  - `distfit()` 함수를 이용해 최적 분포 추정

- **판관비율 (Selling & Admin Expense Ratio)**:  
  - 분포: `Triangular(min, mode, max)`  
    - min: 0.06  
    - mode: 0.07  
    - max: 0.08

- **기타 항목 (고정/변동 값)**:  

| 구분       | 수치            | 비고          |
|------------|-----------------|---------------|
| 월별 감가상각 | 100,000,000원   | 고정금액       |
| 월별 이자비용 | 80,000,000원    | 고정금액       |
| 월별 법인세  | 월별 EBT × 24%  | 변동금액       |
| 발행주식수   | 100,000주       | 발행주식수     |

---

## 결과 변수 산식

- **매출총이익 (Gross Profit)** = 순매출 – 매출원가  
- **EBIT** = 매출총이익 – 판관비 – 감가상각비  
- **EBT** = EBIT – 이자비용  
- **순이익 (Net Income)** = EBT – 법인세  

---

## 시뮬레이션 조건

- **목표**: EPS ≥ 57,000원 달성 확률  
- **Monte Carlo 반복 횟수**: 10,000회  
- **Seed**: `np.random.seed(0)`  
- **라이브러리**: `numpy`, `pandas` 등 활용  
- **출력**: 확률값 (소수점 다섯째 자리에서 반올림, 예: 0.12345 → 0.1235)
